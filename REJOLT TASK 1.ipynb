```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,5))
sns.barplot(x=df['Category'].unique(),y=df.groupby('Category')['Previous Purchases'].sum())
plt.xticks(rotation='vertical')
plt.show()
```
